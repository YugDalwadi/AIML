```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
```

```python
from tensorflow import keras
from keras.datasets import mnist

(xTrain, yTrain), (xTest, yTest) = mnist.load_data()
xTest = xTest.reshape((10000, 784))
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
XtestNew = tsne.fit_transform(xTest)

xTestSubset = XtestNew[:500, :]

plt.scatter(xTestSubset[:, 0], xTestSubset[:, 1], cmap="jet", c=yTest[:500])
# We took the x dim, then the y dim, then set a cmap and let c be any number from 0 to 9
plt.colorbar()
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
XtestNew = pca.fit_transform(xTest)

xTestSubset = XtestNew[:500, :]

plt.scatter(xTestSubset[:, 0], xTestSubset[:, 1], c=yTest[:500], cmap="jet")
plt.colorbar()
plt.show()
```

```python
from sklearn.manifold import LocallyLinearEmbedding

lle = LocallyLinearEmbedding(n_components=2, n_jobs=-1)
XTestNew = lle.fit_transform(xTest)

xTestSubset = XtestNew[:500, :]

plt.scatter(xTestSubset[:, 0], xTestSubset[:, 1], cmap="jet", c=yTest[:500])
plt.colorbar()
plt.show()
```
