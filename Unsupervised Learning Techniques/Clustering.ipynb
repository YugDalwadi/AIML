```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(250, n_features=2, centers=4, random_state=42)

plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
from sklearn.cluster import KMeans

def kMeansTrainer(nFeaturesMax, xSet):
    inertias = []
    for i in range(nFeaturesMax):
        kMeans = KMeans(n_clusters=i+1)
        kMeans.fit(xSet)
        inertias = inertias + [kMeans.inertia_]
    return inertias

nFeatures = 20
InertiaArr = kMeansTrainer(nFeatures, X)
print(InertiaArr)
```

```python
plt.plot([i for i in range(nFeatures)],InertiaArr, "--" )
plt.plot([i for i in range(nFeatures)],InertiaArr, "." )
plt.show()
```
