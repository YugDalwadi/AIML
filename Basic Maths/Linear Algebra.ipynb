```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Vectors can be represented using matrices that go vertically. These matrices can be defined as something like an array in python. However, we shall use ndarrays to make it easier to do mathematical calculations on the values. These vectors can be represented as points or arrows, however we shall try to use arrows in the code below.

```python
# These are two vectors.
a = np.array([1,2])
b = np.array([3,4])

def plot2DVectorArrows(Vector2D, origin=[0,0], **options):
    return plt.arrow(origin[0], origin[1], Vector2D[0], Vector2D[1], head_width=0.2, head_length=0.25, length_includes_head=True, **options)

plot2DVectorArrows(a, color="r")
plot2DVectorArrows(b, color="g")
plt.axis([0,9,0,9]) # Points to draw on axes (First x then y; 2 each)
plt.grid()
plt.show()
```

```python
from matplotlib.patches import Polygon

def plot2DVectorArrows(Vector2D, origin=[0, 0], **options):
    return plt.arrow(origin[0], origin[1], Vector2D[0], Vector2D[1],
              head_width=0.2, head_length=0.3, length_includes_head=True,
              **options)
              
def plotLinTransform(P_before, P_after, text_before, text_after, axis = [0, 5, 0, 4], arrows=False):
    if arrows:
        for vector_before, vector_after in zip(P_before.T, P_after.T):
            plot2DVectorArrows(vector_before, color="blue", linestyle="--")
            plot2DVectorArrows(vector_after, color="red", linestyle="-")
    plt.gca().add_artist(Polygon(P_before.T, alpha=0.2))
    plt.gca().add_artist(Polygon(P_after.T, alpha=0.3, color="r"))
    plt.text(P_before[0].mean(), P_before[1].mean(), text_before, fontsize=18, color="blue")
    plt.text(P_after[0].mean(), P_after[1].mean(), text_after, fontsize=18, color="red")
    plt.axis(axis)
    plt.grid()



```

```python
A = np.array([
    [0,0,1,1], # x
    [0,1,1,0]  # y
])

B_Shear = np.array([
    [1,1], # x
    [0,1]  # y
])
#print(A)
print(B_Shear.dot(A))
plotLinTransform(A, B_Shear.dot(A), "P", r"B_Shear(P)")
```
